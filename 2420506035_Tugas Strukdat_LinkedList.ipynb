```python
class Node:
    def __init__(self, name):
        self.name = name
        self.next = None

class CircularLinkedList:
    def __init__(self):  #
        self.head = None

    def append(self, name):
        new_node = Node(name)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head

    def print_schedule(self, days=14):
        if not self.head:
            print("Jadwal kosong.")
            return

        temp = self.head
        for day in range(1, days + 1):
            print(f"Hari {day}: {temp.name}")
            temp = temp.next  # Perbaikan: hanya pindah satu langkah per hari

# Input nama satpam
satpam_list = CircularLinkedList()
for i in range(4):
    name = input(f"Masukkan nama satpam ke-{i+1}: ")
    satpam_list.append(name)

# Cetak jadwal piket selama 14 hari
print("\nðŸ“… Jadwal Piket Satpam Selama 2 Minggu:\n")
satpam_list.print_schedule(14)
```
